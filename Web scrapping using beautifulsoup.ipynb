# Question2:

 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
# Send the request to the web page server to get the source code of the page:-

Movies=requests.get("https://www.imdb.com/chart/top/?ref_=nv_mv_250")

Movies
```

```python
# Extracting the movies_name from the web source code:-


Movies.content                        # web page content.


soup=BeautifulSoup(Movies.content)    # Using BeautifulSoup for simplification and for Scrapping of data. 

soup


movies=soup.find_all("td",class_="titleColumn") #Extracting the movie name from the soup.

movies

movies_name=[]
for i in movies:
    for j in i.find_all("a"):
        movies_name.append(j.text.replace("\n",""))
movies_name[0:100]
```

```python
imbd=soup.find_all("td",class_="ratingColumn imdbRating") #Extracting the Imdb_rating from the soup.

imbd

imbd_rating=[]     # Printing the imdb_rating.
for i in imbd:
    imbd_rating.append(i.text.replace("\n",""))
imbd_rating[0:100]
```

```python
year=soup.find_all("span",class_="secondaryInfo") # Extracting the Year_relasing date from the soup

year

year_relase=[] # Print the year_relasing date.
for i in year:
    year_relase.append(i.text.replace("\n",""))
year_relase[0:100]
```

```python
# Print the whole information and save into csv formate:-

imbd_movies=pd.DataFrame({})
imbd_movies["Movies_Name"]=movies_name[0:100]
imbd_movies["Imbd_Year_Relase"]=year_relase[0:100]
imbd_movies["Imbd_Ratings"]=imbd_rating[0:100]

imbd_movies
```

```python
# Extracting ODI_Rating.

Rating1=soup.find_all("td",class_="rankings-block__banner--rating u-text-right")
Rating1

Frist_Element_Rating=[]
for i in Rating1:
    Frist_Element_Rating.append(i.text.replace("\n",""))
Frist_Element_Rating

Rating2=soup.find_all("td",class_="table-body__cell u-text-right rating")
Rating2

Rest_Element_Rating=[]
for i in Rating2:
    Rest_Element_Rating.append(i.text.replace("\n",""))
Rest_Element_Rating

Final_Rating=Frist_Element_Rating+Rest_Element_Rating 
Final_Rating[0:10]
```

```python
# Print the whole information and save into csv formate:-

ODI=pd.DataFrame({})
ODI["ODI_TEAM"]=odi_team[0:10]
ODI["ODI_MATCH"]=Final_Match[0:10]
ODI["ODI_POINTS"]=Final_Rest_Element_Points[0:10]
ODI["ODI_RATING"]=Final_Rating[0:10]

ODI
```

# Question8:

Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
url = "https://forecast.weather.gov/MapClick.php?textField1=37.78&textField2=-122.42#.YJJVsLUzZPY"
page12 = requests.get(url)
# see content in page12
soup12 = BeautifulSoup(page12.content)
period = []  #empty list
short_desc = []   #empty list
temperature = []   #empty list
description = []    #empty list

for i in soup12.find_all("div", class_="col-sm-2 forecast-label"):
    period.append(i.text)
for i in soup12.find_all('div',class_="col-sm-10 forecast-text"):        
        short_desc.append(i.text.split(',')[0])
for i in soup12.find_all('div',class_="col-sm-10 forecast-text"):        
        description.append(i.text)
        
for i in soup12.find_all('p',attrs={'short-desc'}):
        if i.next_sibling is not None:
            temperature.append(i.next_sibling.text)
        else:
            temperature.append(' ')
for i in soup12.find_all("div", class_="col-sm-10 forecast-text"):
        description.append(i.text)
# Make data frame
san_francisco_weather=pd.DataFrame({})
san_francisco_weather['Period']=period[:9]
san_francisco_weather['Temperature']=temperature[:9]
san_francisco_weather['Short_description']=short_desc[:9]
san_francisco_weather['Description']=description[:9]
san_francisco_weather
```
