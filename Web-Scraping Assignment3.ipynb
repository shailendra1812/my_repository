# Question:2

2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“?

```python
# Creating empty list for scraping data:

urls=[]
Brand_Name=[]
Rating=[]
Rating_Number=[]
Price=[]
Return=[]
Delivery=[]
Product_Availability=[]
Details=[]

# Scrap the url:

for i in range(0,3):
    for j in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-4']/a"):
        urls.append(j.get_attribute("href"))
    driver.find_element_by_xpath("//li[@class='a-last']/a").click()
    time.sleep(4)
    
# Scrap the Brand_Name:

Brand_Name=[]
for i in urls:
    driver.get(i)
    try:
        brand_name=driver.find_element_by_xpath("//td[@class='a-span9']/span")
        Brand_Name.append(brand_name.text)
    except NoSuchElementException:
        Brand_Name.append("-") 

# Scrap the Rating:

Rating=[]
for i in urls:
    driver.get(i)
    try:
        rating=driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']/span")
        Rating.append(rating.text)
    except NoSuchElementException:
        Rating.append("-")

# Scrap No of Rating:

Rating_Number=[]
for i in urls:
    driver.get(i)
    try:
        rating_number=driver.find_element_by_id("acrCustomerReviewLink")
        Rating_Number.append(rating_number.text)
    except NoSuchElementException:
        Rating_Number.append("-")

# Scrap the Price:

Price=[]
for i in urls:
    driver.get(i)
    try:
        price=driver.find_element_by_id("priceblock_ourprice")
        Price.append(price.text)
    except NoSuchElementException:
        Price.append("-")
        
# Scrap the Return/Exchange:

Return=[]
for i in urls:
    driver.get(i)
    try:
        Exchange=driver.find_element_by_xpath("//div[@class='a-section a-spacing-none icon-content']/a")
        Return.append(Exchange.text)
    except NoSuchElementException:
        Return.append("-")
        
# Scrap Excepted Delivery:

Delivery=[]
for i in urls:
    driver.get(i)
    try:
        delivery=driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[5]/div[4]/div[4]/div[19]/div[2]/div/div/div/div[2]/div[2]/a")
        Delivery.append(delivery.text)
    except NoSuchElementException:
        Delivery.append("-")


# Scrap the Availability

Product_Availability=[]
for i in urls:
    driver.get(i)
    try:
        availability=driver.find_element_by_id("availability")
        Product_Availability.append(availability.text)
    except NoSuchElementException:
        Product_Availability.append("-")
        
# Scrap Other Detail:

Details=[]
for i in urls:
    driver.get(i)
    try:
        details=driver.find_element_by_id("productTitle")
        Details.append(details.text)
    except NoSuchElementException:
        Details.append("-")
        
# Creating the DataFrame of Scraped data and taking only # page of search reasult:

Data_Frame=pd.DataFrame({"Brand_Name":Brand_Name,"Ratings":Rating,"No. of Ratings":Rating_Number,"Price":Price,
                        "Return/Exchange":Return,"Expected Delivery":Delivery,"Availability":Product_Availability,
                        "Other Details":Details,"Product_URL":urls})
Data_Frame

# Save the dataFrame into csv formate:

Data_Frame.to_csv("Amazon_Product1")

# Close the Driver:

driver.close()
```

# Question:4

4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV?

```python
# Connect to the webdriver:

driver=webdriver.Chrome("chromedriver.exe")

driver.get("https://www.flipkart.com/")   #Search url
time.sleep(4)

# Get the search button and the keys "pixel 4A" :

search_product=driver.find_element_by_xpath("//input[@class='_3704LK']")
search_product.send_keys("pixel 4A")
time.sleep(4)

# Click the search button:

search_button=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_button.click()
time.sleep(4)
```

```python
# Creating empty list for scraping data:

Brand_Name=[]
Color=[]
RAM=[]
Storage_ROM=[]
Primary_Camera=[]
Secondary_Camera=[]
Display_Size=[]
Display_Resolution=[]
Processor=[]
Processor_Cores=[]
Battery_Capacity=[]
Price=[]

# Scrap the url:

urls=[]
for i in driver.find_elements_by_xpath("//a[@class='_1fQZEK']"):
    urls.append(i.get_attribute("href"))

# Scrap the Brand_Name:
    
Brand_Name=[]
for i in urls:
    driver.get(i)
    try:
        brand=driver.find_element_by_xpath(" //div[@class='_3k-BhJ'][1]/table/tbody/tr[3]/td[2]/ul/li")
        Brand_Name.append(brand.text)
    except NoSuchElementException:
        Brand_Name.append("-")
        
        
# Scrap the camera:        
Color=[]
for i in urls:
    driver.get(i)
    try:
        color=driver.find_element_by_xpath("//div[@class='_3k-BhJ'][1]/table/tbody/tr[4]/td[2]/ul/li")
        Color.append(color.text)
    except NoSuchElementException:
        Color.append("-")

    

# Scrap the Storage:        

Storage=[]
for i in urls:
    driver.get(i)
    try:
        ram=driver.find_element_by_xpath("//li[@class='_21Ahn-']")
        Storage.append(ram.text.replace("|",","))
    except NoSuchElementException:
        Storage.append("-")

        
# Scrap RAM:

RAM=[]
for i in range(0,len(Storage)):
    RAM.append(Storage[i][0:8])

    
#  Scrap ROM:

Storage_ROM=[]
for i in range(0,len(Storage)):
    Storage_ROM.append(Storage[i][10:21])

    
    
# Scraping Camera of the phone:

Camera=[]
for i in urls:
    driver.get(i)
    try:
        cam=driver.find_element_by_xpath("//li[@class='_21Ahn-'][3]")
        Camera.append(cam.text.replace("|",","))
    except NoSuchElementException:
        Camera.append("-")
        
        
# Primary_Camera:(Rarer Camera)

Primary_Camera=[]
for i in range(0,len(Camera)):
    Primary_Camera.append(Camera[i][0:17])
    
    
# Secondary_Camera:(Front camera) 

Secondary_Camera=[]
for i in range(0,len(Camera)):
    if Camera[i][19:42]=="":
        Secondary_Camera.append("-")
    else:
        Secondary_Camera.append(Camera[i][19:42])
        
# Scrap Display_Size:

Display=[]
for i in urls:
    driver.get(i)
    try:
        size=driver.find_element_by_xpath("//li[@class='_21Ahn-'][2]")
        Display.append(size.text)
    except NoSuchElementException:
        Display.append("-")

# Display_Size:

Display_Size=[]
for i in range(0,len(Display)):
    Display_Size.append(Display[i][0:20])
    
    
    
# Display_Resolution:
Display_Resolution=[]
for i in range(0,len(Display)):
    Display_Resolution.append(Display[i][21:37])
    

# Scrap Processor:

Processor=[]
for i in urls:
    driver.get(i)
    try:
        r=driver.find_element_by_xpath("//li[@class='_21Ahn-'][5]")
        Processor.append(r.text)
    except NoSuchElementException:
        Processor.append("-")
        
        
        
# Scrap Processor_Cores:

Processor_Cores=[]
a="Octa-Core"
for i in range(0,24):
    Processor_Cores.append(a)
    
    

# Scrap Battery_Capacity:

Battery_Capacity=[]
for i in urls:
    driver.get(i)
    try:
        b=driver.find_element_by_xpath("//li[@class='_21Ahn-'][4]")
        Battery_Capacity.append(b.text)
    except NoSuchElementException:
        Battery_Capacity.append("-")
        

        
# Scrape Price:

Price=[]
for i in urls:
    driver.get(i)
    try:
        p=driver.find_element_by_xpath("//div[@class='_30jeq3 _16Jk6d']")
        Price.append(p.text)
    except NoSuchElementException:
        Price.append("-")        
```

```python
# Make the DataFrame:

DataFrame=pd.DataFrame({"Brand_Name":Brand_Name,"Color":Color,"RAM":RAM,"Storage_ROM":Storage_ROM,
                       "Primary_Camera":Primary_Camera,"Secondary_Camera":Secondary_Camera,"Display_Size":Display_Size,
                       "Processor":Processor,"Processor_Cores":Processor_Cores,"Battery_Capacity":Battery_Capacity,
                       "Price":Price,"Display_Resolution":Display_Resolution,"urls":urls})
DataFrame

# Save the dataframe into csv formate:

DataFrame.to_csv("Flipkart_Moble_Detail")

# Close the driver:

driver.close()
```
